# Option Stratagies

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_payoff_diagram(spot_price, payoff, strategy):
    plt.xlabel("Spot Price")
    plt.ylabel("Payoff")
    plt.title(strategy)
    plt.plot(spot_price, payoff)
```

## Bull Call Spread

```python
def bull_call_spread(lower_bound, upper_bound, strike_price_long, strike_price_short, \
                     option_premium_long, option_premium_short):
    spot_price = np.arange(lower_bound, upper_bound)
    payoff = (np.maximum(spot_price - strike_price_long, 0) - option_premium_long) + \
             (option_premium_short - np.maximum(spot_price - strike_price_short, 0))
    return spot_price, payoff
```

```python
spot_price, payoff = bull_call_spread(80, 180, 120, 140, 5, 2)
plot_payoff_diagram(spot_price, payoff, "Bull Call Spread")
```

## Short Straddle

```python
def short_straddle(lower_bound, upper_bound, strike_price, option_premium_long, option_premium_short):
    spot_price = np.arange(lower_bound, upper_bound)
    payoff = (option_premium_long - np.maximum(spot_price - strike_price, 0)) + \
         (option_premium_short - np.maximum(strike_price - spot_price, 0))
    return spot_price, payoff
```

```python
spot_price, payoff = short_straddle(7200, 8000, 7600, 77, 88)
plot_payoff_diagram(spot_price, payoff, "Short Straddle")
```
